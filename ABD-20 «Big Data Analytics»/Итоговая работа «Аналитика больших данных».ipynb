## 1. Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы.

```
import pandas as pd
```

```
columns = ['user_id', 'item_id', 'rating', 'timestamp']
ratings = pd.read_csv('u.data.csv', sep='\s+', header = None, names = columns)
ratings.head()
```

```
columns_u = ['movie_id', 'movie_title', 'release_date', 'video_release_date',
             'imdb_url', 'unknown', 'action', 'adventure', 'animation', 'childrens',
             'comedy', 'crime', 'documentary', 'drama', 'fantasy', 'fil_noir',
             'horror', 'musical', 'mystery', 'romance', 'sci-fi', 'thriller',
             'war', 'western']
movies = pd.read_csv('u.item.csv', sep='|', encoding = 'windows-1251', header = 1, names = columns_u)
movies.head()
```

## 2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```
ratings[['user_id','item_id']].groupby('user_id').count().sort_values('item_id', ascending = False).head(1)
```

## 3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```
ratings_u = ratings[(ratings['user_id'] > 404) & (ratings['user_id'] < 406)]
ratings_u.head()
```

## 4. Добавьте к датафрейму из задания 3 столбцы:
- По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет
- столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```
genres = ratings_u.merge(movies, left_on='item_id', right_on='movie_id')
genres.head()
```

```
avg_rating = ratings[['item_id','rating']].groupby('item_id').mean()
number_ratings = ratings[['item_id','rating']].groupby('item_id').count()
all = number_ratings.merge(avg_rating, left_on='item_id', right_on='item_id')
all.head()
```

```
user_405 = genres.merge(all, left_on='item_id', right_on='item_id')
user_405.head()
```

## 5. Сформируйте X_train, X_test, y_train, y_test

```
X = user_405[['unknown', 'action', 'adventure', 'animation', 'childrens',
             'comedy', 'crime', 'documentary', 'drama', 'fantasy', 'fil_noir',
             'horror', 'musical', 'mystery', 'romance', 'sci-fi', 'thriller',
             'war', 'western', 'rating_x', 'rating_y']]
Y = user_405['rating']
```

## 6. Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

## 7. Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```
lr.coef_
```

```
mean_squared_error(Y_test, lr.predict(X_test))
```

```
mean_squared_error(Y_train, lr.predict(X_train))
```

### Загрузка данных при помощи spark.read.csv

```
df = spark.read.csv('u.data.csv', inferSchema=True, header=False, sep='\t')
df = df.withColumnRenamed('_c0','user_id').withColumnRenamed('_c1','movie_id').withColumnRenamed('_c2','rating').withColumnRenamed('_c3','timestamp')
df.show()
```

```
df2 = spark.read.csv('u.item.csv', inferSchema=True, header=True, sep= '|')
df2.show()
```

## 9. Средствами спарка вывести среднюю оценку для каждого фильма

```
from pyspark.sql.functions import *
```

```
ratings = df.join(df2, df[1] == df2[0], 'left')
ratings.show()
```

```
ratings = ratings.drop('movie_id', 'video_release_date', 'IMDb_URL', 'timestamp')
ratings.show()
```

```
x = ratings.groupBy('movie id', 'movie title').avg('rating')
x.show()
```

## 10. В спарке получить 2 датафрейма с 5-ю самыми популярными и самыми непопулярными фильмами

```
x.orderBy('avg(rating)', ascending=False).show(5)
```

```
x.orderBy('avg(rating)', ascending=True).show(5)
```
