```python
import pandas as pd
```

```python
df_csv = pd.read_csv('analytics.csv', sep=',')
df_csv.info()
```

```python
df_sql = pd.read_sql(
"""
SELECT 
  *
FROM
  city
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_sql.head()
```

```python
city_csv = df_csv['Транслит'].tolist()
city_csv
```

Поиск городов в City и запись в датафрейм

```python
df_city = df_sql[df_sql['name'].isin(city_csv)]
df_city
```

Соединение двух датафреймов

```python
df_merged = df_csv.merge(df_city, left_on='Транслит', right_on='name', how='inner')
df_merged
```
