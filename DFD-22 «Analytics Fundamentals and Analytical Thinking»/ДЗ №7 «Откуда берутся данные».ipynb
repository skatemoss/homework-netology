Соединение таблиц с помощью SQL

```python
df = pd.read_sql(
"""
SELECT 
  *
FROM
  country ct
INNER JOIN
  country_language ctl
ON
  ct.code = ctl.country_code
WHERE
  is_official = 'True'
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df
```

Соединение таблиц с помощью команды Merge

```python
df_country_language = pd.read_sql(
"""
SELECT 
  *
FROM
  country_language
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_country_language.head()
```

```python
df_country = pd.read_sql(
"""
SELECT 
  *
FROM
  country
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
df_country.head()
```

```python
df_merged = df_country.merge(df_country_language, left_on='code', right_on='country_code', how='inner')
df_merged.query('is_official == True')
```
