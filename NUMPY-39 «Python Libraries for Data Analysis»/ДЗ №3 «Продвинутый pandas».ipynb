### Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по правилам:

    если источник traffic_source равен Yandex или Google, то в source_type ставится organic;
    для источников paid и email из России ставим ad;
    для источников paid и email не из России ставим other;
    все остальные варианты берём из traffic_source без изменений.


```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/homework/visit_log.csv', sep=';')
df.head()
```

```python
def source_type(df):
  
  if df['traffic_source'] == 'yandex' or df['traffic_source'] == 'google':
    df['source_type'] = 'organic'
  elif ((df['traffic_source'] == 'paid' or df['traffic_source'] == 'email') and df['region'] == 'Russia'):
    df['source_type'] = 'ad'
  elif ((df['traffic_source'] == 'paid' or df['traffic_source'] == 'email') and df['region'] != 'Russia'):
    df['source_type'] = 'other'
  else:
    df['source_type'] = df['traffic_source']
        
  return df['source_type']
```

```python
df['source_type'] = df.apply(source_type, axis = 1)
```

### Задание 3
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
df = pd.read_csv('/content/drive/MyDrive/ml-latest-small/ratings.csv')
df.head(10)
```

```python
def diff(timestamp):
    return max(timestamp) - min(timestamp)
```

```python
df = pd.pivot_table(df, index = 'userId', values = ['rating', 'timestamp'], aggfunc = {'rating':['mean', 'count'], 'timestamp':['min', 'max']})
```

```python
df = df[df['rating', 'count'] > 100]
df.head(10)
```

```python
df['life'] = df.apply(diff, axis = 1)
df.head(10)
```
