### Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
df = pd.read_csv('/content/drive/MyDrive/power.csv')
df.head()
```

```python
df = df[(df['country'].isin(['Latvia', 'Estonia', 'Lithuania']))
        & (df['quantity'] >= 0)
        & ((df['year'] > 2005) & (df['year'] < 2010))
        & ((df['category'] == 4) | (df['category'] == 12) | (df['category'] == 21))]
df['quantity'].sum()
```
