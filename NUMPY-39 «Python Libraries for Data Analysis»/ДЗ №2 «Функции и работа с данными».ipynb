###Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

    оценка 2 и ниже — низкий рейтинг;
    оценка 4 и ниже — средний рейтинг;
    оценка 4.5 и 5 — высокий рейтинг.

Результат классификации запишите в столбец class.

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ml-latest-small/ratings.csv')
df_movies = pd.read_csv('/content/drive/MyDrive/ml-latest-small/movies.csv')
df = df.merge(df_movies, left_on='movieId', right_on='movieId', how='inner')[['movieId', 'title', 'rating']]
df.head()
```

```python
df_rating = pd.pivot_table(df, index=['movieId', 'title'], values='rating', aggfunc='mean').reset_index()
df_rating.head()
```

```python
def movie_class(rating):
  if rating <= 2:
    rating_class = 'Низкий рейтинг'
  elif rating >= 4.5:
    rating_class = 'Высокий рейтинг'
  else:
    rating_class = 'Средний рейтинг'
  return rating_class
```

```python
df_rating['class'] = df_rating['rating'].apply(movie_class)
df_rating.head()
```
